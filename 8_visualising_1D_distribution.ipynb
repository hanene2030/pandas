# Visualising-1D-distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
df = pd.read_csv("heart.csv")
df.head(2)
```

## 1.Historgrams
### Most common to show distribution

```python
df.age.plot.hist(bins=30);
```

```python
plt.hist(df.age, bins=30);
```

```python
fig,ax = plt.subplots()
ax.hist(df.trestbps, bins=30, histtype="step" , label="trestbps")
ax.hist(df.thalach, bins=30, histtype="stepfilled" , label="thalach", alpha=0.3 , edgecolor="red")
ax.legend();
```

## 2.Box plot
### shows mean, quantiles, IQR and outliers

```python
df[["trestbps","thalach"]].plot.box()
```

```python
df[["trestbps","thalach"]].describe()
```

```python
plt.boxplot(df[["trestbps","thalach"]].to_numpy());
```

```python
sb.boxplot(data=df[["trestbps","thalach"]]);
```

```python
sb.boxplot(x="cp", y="trestbps",data=df);
```

```python
df.groupby("cp").boxplot(column="trestbps", sharex=True, layout=(1,4), grid=False)
plt.tight_layout();
```

## 3. Violin plots
### Like box plots, but better

```python
fix, ax = plt.subplots()
ax.violinplot(df[["trestbps","thalach"]].to_numpy());
```

```python
fix, ax = plt.subplots()
ax.violinplot(df[["trestbps","thalach"]].to_numpy(), bw_method=0.2);
```

```python
sb.violinplot(data=df[["trestbps","thalach"]], inner="quartile", bw=0.2);
```

## 4.Bee swarm plots


```python
sb.swarmplot(data=df[["trestbps","thalach"]], size=3);
```

```python
sb.violinplot(data=df[["trestbps","thalach"]], inner=None);
sb.swarmplot(data=df[["trestbps","thalach"]], size=2, color="r", alpha=0.5);
```
