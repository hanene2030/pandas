# Merging basic


```python
import pandas as pd
```

```python
students = pd.read_csv("students.csv")
teachers = pd.read_csv("teachers.csv")
grades2 = pd.read_csv("grades2.csv")
grades1 =  pd.read_csv("grades1.csv")
contacts = pd.read_csv("contact.csv")
```

## Concat and append

```python
grades1["Semester"] = 1
grades2["Semester"] = 2
```

```python
grades1.head(2)
```

```python
grades = pd.concat([grades1, grades2])
grades.sample(3)
```

```python
grades1.copy().append(grades2)
```

```python
#Joining the grades together isnt updating the index
grades = grades.reset_index(drop=True)
grades
```

## Merging horizontlly

```python
students_grades = pd.merge(students, grades, left_on="id", right_on="student_id")
students_grades
```

```python
print(students.shape, grades.shape, students_grades.shape)
#This shows that we havent dropped anything in the students grades
```

```python
students2 = students.rename({"id" : "student_id"}, axis=1)
students_grades2 = pd.merge(students2, grades, on="student_id")
students_grades2
```

```python
print(students2.shape, grades.shape, students_grades2.shape)
```

```python
students_full = students_grades2.merge(contacts , on="student_id")
students_full.head(3)
```

```python
print(students_grades2.shape, contacts.shape, students_full.shape)
```

```python
result = pd.merge(teachers,students_full , on="course" )
result.head(4)
```

```python
result[result.grade == "A"].head(3)
```

```python
students_f = result.loc[result.grade == "F", "student_id" ]
students_f.head(3)
```

```python
# number of classes that every students is failling
method1 = students_f.value_counts().rename_axis("student_id").reset_index(name="counts")
method1.head(3)
```

```python
method2 = result.loc[result.grade == "F",[ "student_id" , "grade"]].groupby("student_id").count().reset_index()
method2.columns = ["student_id", "counts"]
method2.head(3)
```

```python
spb = method2[method2.counts > 3].student_id
spb.head()
```

```python
final = students_full[students_full.student_id.isin(spb)]
final.parent_contact.unique()
```
