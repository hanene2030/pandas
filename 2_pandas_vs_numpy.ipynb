# Pandas vs numpy

1. Pandas has a numpy core
2. They do present different data structures

```python
import pandas as pd
import numpy as np
df = pd.read_csv("heart.csv")
df.head(2)
```

```python
data = df.to_numpy()
```

```python
print(data.dtype,data) # First thing to note, the data type is float 64. numpy converted our entire array to a single type floats 
```

```python
data[0,0] =100
data, df.head(2) # the numpy data has changed, but the original panda's data frame not
```

```python
df2 = df[["age","sex","cp"]]
data2 =  df2.to_numpy()
data2[0,0] =100
df2.head(2) 
# just be aware that even if you're working with a non-pandas object,still very likely be modifying the panda's
```

```python
df3 = df[["age","sex","cp"]]
#data3 =  df3.to_numpy(copy=True)
data3 =  df3.to_numpy().copy()
data3[0,0] =100
df3.head(2) 
```
