# Ploting methods

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("heart.csv")
df.head(2)
```

```python
fig,axes = plt.subplots(ncols=2)
df.plot.scatter(x="age",y="chol", ax= axes[0])
# trestbps = resting blood pressure
df.plot.scatter(x="age",y="trestbps", ax= axes[1])
fig.tight_layout();
```

### To save out your plots

```python
fig2 = axes[0].get_figure()
fig2.savefig("plot.png", bbox_inches="tight", tranparent=True)
```

```python
with plt.style.context("dark_background"):
  fig,axes = plt.subplots(ncols=3, sharey=True,
                          gridspec_kw={"width_ratios":[1,2,1], "wspace":0})
  y = "age"
  xs = ["chol", "trestbps", "thalach"]
  for x , ax in zip(xs, axes):
    ax.scatter(df[x],df[y])
    ax.set_xlabel(x)
  axes[0].set_ylabel(y);
```

```python
with plt.style.context("seaborn"):
  fig,axes = plt.subplots(ncols=3, sharey=True,
                          gridspec_kw={"width_ratios":[1,2,1], "wspace":0})
  y = "age"
  xs = ["chol", "trestbps", "thalach"]
  for x , ax in zip(xs, axes):
    ax.scatter(df[x],df[y])
    ax.set_xlabel(x)
  axes[0].set_ylabel(y);
```

```python
with plt.style.context("default"):
  fig,axes = plt.subplots(ncols=3, sharey=True,
                          gridspec_kw={"width_ratios":[1,2,1], "wspace":0})
  y = "age"
  xs = ["chol", "trestbps", "thalach"]
  for x , ax in zip(xs, axes):
    ax.scatter(df[x],df[y])
    ax.set_xlabel(x)
  axes[0].set_ylabel(y);
```

```python
fig.savefig("output.png", dpi=100, bbox_inches="tight")
fig.savefig("output.pdf", dpi=100, bbox_inches="tight")
```
