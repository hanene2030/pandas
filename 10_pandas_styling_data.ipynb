# Pandas styling data

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(data=np.random.normal(size=(6,6)), columns=[x for x in "ABCDEF"] )
df
```

```python
#Show all negative numbers in red
def neg_red(x):
  return f"color:{'red' if x < 0 else 'green'}"
df.style.applymap(neg_red)
```

```python
def gold_max(xs):
  m = xs.to_numpy().max()
  color ={True: "background-color: yellow",False: ""}
  is_max = (xs == m).replace(color)
  return is_max
df.style.apply(gold_max, axis=1) # row by row
```

```python
df.style.background_gradient(cmap="magma")
```

```python
df.abs().style.bar(align="left", width=95)
```

```python
df.abs().style.highlight_min(["A","B"], color="red")
```

```python
df.abs().style.highlight_between(color='blue', axis=None, left=1, right=2, inclusive='both')
```

```python
df.style.text_gradient(cmap='inferno') # Matplotlib colormap
```

```python
head = [{"selector": "th", "props": [("text-align", "center")]}]
df.style.set_table_styles(head).bar(align='mid', color=["blue", "green"], vmin=3, vmax=3)
```

```python
# we can chain them all together
df.style.bar(align="mid", width=50).apply(gold_max, axis=None).applymap(neg_red)
```

```python
df.style.text_gradient(cmap='PuBu')
```
